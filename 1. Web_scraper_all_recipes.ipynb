```python
import requests
import bs4
from urllib.request import urlopen
import re
import pandas as pd
import numpy as np
```

## Check if site exists

```python
list_of_links = []
```

```python
for n in range(267000,267200):
    link_test = 'https://www.allrecipes.com/recipe/{}/'.format(n)
    res1 = requests.get(link_test)
    soup = bs4.BeautifulSoup(res1.text, 'lxml')
    # If check is null then the page doesn't exist
    check = soup.find_all('h2')
    error = soup.select('.error-page')
    if check == []:
        continue
    elif check[5].text == 'Inspiration and Ideas':
        continue
    else:
        list_of_links.append(link_test)
```

## Preparing data frames

```python
ingredients_table = pd.DataFrame({})
nutr_table = pd.DataFrame({})
```

## For loop for data extract

```python
for item in range(len(list_of_links)):
    res1 = requests.get(list_of_links[item])
    soup = bs4.BeautifulSoup(res1.text, 'lxml')
    
    title = soup.select('.headline-wrapper')
    title = title[0].text
    title = title.strip()
    
    ingredients = soup.select('.ingredients-item-name')
    for n in range(len(ingredients)):
        ingredients[n] = ingredients[n].text
        ingredients[n] = ingredients[n].strip()
    ingredients.insert(0, title)
    #ingredients_table.loc[df_length] = ingredients
    
    ingredients_table = ingredients_table.append([ingredients], ignore_index = True)
    
    
    nutrition = soup.select('.nutrient-name')
    nutr_dictionairy = {}
    for n in range(len(nutrition)):
        nutr = nutrition[n].text
        nutr_name = nutr.split(':')[0]
        value = nutr.split(':')[1]
        value = value.strip()
        nutr_dictionairy['meal'] = title
        nutr_dictionairy['link'] = list_of_links[item]
        nutr_dictionairy[nutr_name] = value
        
    nutr_table = nutr_table.append(nutr_dictionairy, ignore_index=True)
```

## Saving dataframes to CSV format
